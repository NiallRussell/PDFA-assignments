```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("https://drive.google.com/uc?id=1AWPf-pJodJKeHsARQK_RHiNsE8fjPCVK&export=download")
```

```python
df
```

```python
df["domain"] = df["Email"].str.split("@").str[1]
df.head(3)
```

```python
domain_count = df["domain"].value_counts()
domain_count
```

```python
fig, ax = plt.subplots()
labels = "example.org", "example.com", "example.net"
ax.pie(domain_count, labels=labels, autopct='%1.1f%%', colors = ["skyblue", "lightpink", "lightgreen"])
ax.set_title('Domain names')
```
